```python
columnas = {
    'customerID': 'id',
    'customer.gender': 'genero',
    'customer.SeniorCitizen': 'tiene +60',
    'customer.Partner': 'posee_pareja',
    'customer.Dependents': 'posee_dependientes',
    'customer.tenure': 'tiempo_contrato',
    'phone.PhoneService': 'servicio_telefono',
    'phone.MultipleLines': 'multiples_lineas',
    'internet.InternetService': 'tipo_internet',
    'internet.OnlineSecurity': 'seguridad_online',
    'internet.OnlineBackup': 'backup_online',
    'internet.DeviceProtection': 'proteccion_dispositivo',
    'internet.TechSupport': 'soporte_tecnico',
    'internet.StreamingTV': 'streaming_tv',
    'internet.StreamingMovies': 'streaming_peliculas',
    'account.Contract': 'tipo_contrato',
    'account.PaperlessBilling': 'factura_digital',
    'account.PaymentMethod': 'metodo_pago',
    'account.Charges.Monthly': 'valor_mensual',
    'account.Charges.Total': 'total_cobrado'
}
df = df.rename(columns= columnas)
df.head()
```

```python
df['genero'] = df['genero'].replace({'Female': 'Femenino', 'Male': 'Masculino'})
df['multiples_lineas'] = df['multiples_lineas'].replace({'No phone service': 'Sin servicio de teléfono'})
df['tipo_internet'] = df['tipo_internet'].replace({'No': '0', 'Fiber optic': 'Fibra óptica'})
df['seguridad_online'] = df['seguridad_online'].replace({'No internet service': 'Sin servicio de internet'})
df['backup_online'] = df['backup_online'].replace({'No internet service': 'Sin servicio de internet'})
df['proteccion_dispositivo'] = df['proteccion_dispositivo'].replace({'No internet service': 'Sin servicio de internet'})
df['soporte_tecnico'] = df['soporte_tecnico'].replace({'No internet service': 'Sin servicio de internet'})
df['streaming_tv'] = df['streaming_tv'].replace({'No internet service': 'Sin servicio de internet'})
df['streaming_peliculas'] = df['streaming_peliculas'].replace({'No internet service': 'Sin servicio de internet'})
df['tipo_contrato'] = df['tipo_contrato'].replace({'One year': 'Anual', 'Month-to-month': 'Mensual', 'Two year': 'Bienal'})
df.head()
```

```python
df['metodo_pago'] = df['metodo_pago'].replace({
    'Mailed check': 'Cheque enviado por correo',
    'Electronic check': 'Cheque electrónico',
    'Credit card (automatic)': 'Tarjeta de crédito (automático)',
    'Bank transfer (automatic)': 'Transferencia bancaria (automática)'
})
df.head()
```

```python
binarios = [
    'Churn',
    'posee_pareja',
    'posee_dependientes',
    'servicio_telefono',
    'multiples_lineas',
    'seguridad_online',
    'backup_online',
    'proteccion_dispositivo',
    'soporte_tecnico',
    'streaming_tv',
    'streaming_peliculas',
    'factura_digital'
]

for col in binarios:
    df[col] = df[col].map({
        'Yes': 1,
        'No': 0
        })
df.head()
```

```python
describe = df.describe()

formatted = describe.astype(object)

binarias = [
    'Churn', 'tiene +60', 'posee_pareja', 'posee_dependientes',
    'servicio_telefono', 'multiples_lineas', 'tipo_internet', 'seguridad_online',
    'backup_online', 'proteccion_dispositivo', 'soporte_tecnico',
    'streaming_tv', 'streaming_peliculas', 'factura_digital'
]

for col in binarias:
    if col in describe.columns:
        for idx in formatted.index:
            if idx == 'count':
                formatted.loc[idx, col] = f"{int(describe.loc[idx, col])}"
            else:
                formatted.loc[idx, col] = f"{describe.loc[idx, col]*100:.1f}%"

monetarias = ['valor_mensual', 'total_cobrado', 'cuentas_diarias']
for col in monetarias:
    if col in describe.columns:
        for idx in formatted.index:
            if idx == 'count':
                formatted.loc[idx, col] = f"{int(describe.loc[idx, col])}"
            else:
                formatted.loc[idx, col] = f"${describe.loc[idx, col]:,.1f}"

contrato = 'tiempo_contrato'
if contrato in describe.columns:
    for idx in formatted.index:
        if idx == 'count':
            formatted.loc[idx, contrato] = f"{int(describe.loc[idx, contrato])}"
        else:
            formatted.loc[idx, contrato] = f"{describe.loc[idx, contrato]:.1f}"

display(formatted)

print("\nNotas interpretativas:")
print(f"Churn: {describe.loc['mean','Churn']*100:.1f}% de clientes cancelaron.")
print(f"Tiene +60: {describe.loc['mean','tiene +60']*100:.1f}% son adultos mayores.")
print(f"Posee pareja: {describe.loc['mean','posee_pareja']*100:.1f}% tienen pareja.")
print(f"Posee dependientes: {describe.loc['mean','posee_dependientes']*100:.1f}% tienen dependientes.")
print(f"Tiempo de contrato: mediana = {describe.loc['50%','tiempo_contrato']:.1f} meses, máximo = {describe.loc['max','tiempo_contrato']:.1f} meses.")
print(f"Valor mensual: media = ${describe.loc['mean','valor_mensual']:.1f}, mediana = ${describe.loc['50%','valor_mensual']:.1f}.")
print(f"Total cobrado: media = ${describe.loc['mean','total_cobrado']:.1f}, mediana = ${describe.loc['50%','total_cobrado']:.1f}, máximo = ${describe.loc['max','total_cobrado']:.1f}.")
print(f"Cuentas diarias: media = ${describe.loc['mean','cuentas_diarias']:.1f}, mediana = ${describe.loc['50%','cuentas_diarias']:.1f}.")
```

```python

df_grouped = df.groupby(['tiempo_contrato', 'Churn']).size().unstack(fill_value=0)

df_grouped['churn_rate'] = (df_grouped.get(1, 0) / df_grouped.sum(axis=1)) * 100

print(df_grouped['churn_rate'].map(lambda x: f'{x:.1f}%').to_string())
```

```python
fig = px.line(df_grouped.reset_index(),
              x='tiempo_contrato',
              y='churn_rate',
              markers=True,
              title='Tasa de Churn (%) por tiempo_contrato',
              color_discrete_sequence=['purple']
             )

fig.update_layout(title_x=0.5)
fig.update_traces(hovertemplate='tiempo_contrato=%{x}tasa de Churn=%{y:.1f}%')
fig.update_xaxes(title_text='tiempo_contrato (meses)')
fig.show()


df_corr = df.copy()


for col in df_corr.select_dtypes(include=['object']).columns:
    if col != 'id':
        df_corr[col] = df_corr[col].astype('category').cat.codes


df_corr = df_corr.rename(columns={'Churn': 'Evasion'})


correlation_matrix = df_corr.drop(columns=['id']).corr()


plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación')
plt.show()


print(correlation_matrix['Evasion'].sort_values(ascending=False).round(2).to_string())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_corr_processed = df.copy()
for col in df_corr_processed.select_dtypes(include=['object']).columns:
    if col != 'id':
        df_corr_processed[col] = df_corr_processed[col].astype('category').cat.codes

df_corr_processed = df_corr_processed.drop(columns=['id'])

plt.figure(figsize=(16, 9))
sns.set_style("whitegrid")

correlaciones = df_corr_processed.corr()['Churn'].sort_values(ascending=False)
colores = sns.color_palette("viridis", len(correlaciones))

ax = correlaciones.plot(kind='bar', color=colores, edgecolor='black', alpha=0.8)

for p in ax.patches:
    height = p.get_height()
    va_position = 'bottom' if height > 0 else 'top'
    offset = 7 if height > 0 else -18

    ax.annotate(f'{height:.2f}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center',
                va=va_position,
                xytext=(0, offset),
                textcoords='offset points',
                fontsize=10,
                fontweight='bold')

plt.title('Correlación de datos con la tasa de abandono (Churn)', fontsize=18, pad=20, fontweight='bold')
plt.ylabel('Coeficiente de correlación', fontsize=13)
plt.xlabel('Variables', fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.axhline(0, color='black', linewidth=1.2)
plt.grid(axis='y', linestyle=':', alpha=0.7)

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

**Introducción**

comprender los factores críticos que impulsan la deserción de clientes (Churn), la cual se sitúa en un 26.5% - 26.6%. El objetivo central es examinar las métricas demográficas, contractuales y de servicio para identificar patrones de comportamiento. A través de este estudio, se busca no solo limpiar e interpretar la información histórica, sino generar insights accionables que permitan priorizar segmentos de alto riesgo, reducir la pérdida de ingresos y sentar las bases para futuros modelos predictivos que garanticen la sostenibilidad del negocio.

**Limpieza y Tratamiento de Datos**

El proceso de preparación de la información fue fundamental para asegurar la validez de las conclusiones estadísticas. Se llevaron a cabo los siguientes pasos:

Extracción y Carga: Los datos fueron obtenidos en formato JSON desde repositorios remotos a través de una API, siendo convertidos a un DataFrame de Pandas para su manipulación estructurada.

Gestión de Inconsistencias: Se realizó una auditoría de valores nulos y vacíos, detectando 224 casos de Churn sin registrar y 11 valores faltantes en la columna de cargos totales, los cuales fueron tratados para no sesgar el análisis.

Traducción y Estandarización: Se renombraron las columnas al español (ej. de tenure a tiempo_contrato) y se tradujeron los registros categóricos (género, métodos de pago, tipos de servicio).

Ingeniería de Variables:

Se transformaron variables binarias de texto ("Sí"/"No") a formato numérico (1/0).

Se utilizó limpieza con expresiones regulares para convertir los cargos totales a tipo de dato flotante.

Se creó la métrica "cuentas_diarias" (mensualidad dividida por 30) para entender el impacto del gasto diario en la percepción del cliente.

Se aplicó codificación categórica (cat.codes) para permitir el cálculo de matrices de correlación.

**Análisis Exploratorio de Datos (EDA)**

Mediante visualizaciones avanzadas y análisis de distribución, se identificaron tendencias determinantes:

Comportamiento del Tiempo (Tenure): La mediana de antigüedad de los clientes es de 29 meses. Sin embargo, existe una correlación negativa moderada entre la antigüedad y el Churn: a mayor tiempo en la empresa, menor es la probabilidad de abandono.

Segmentación del Riesgo Contractual: El gráfico de tipos de contrato revela que los clientes con suscripción mes a mes tienen una tasa de deserción del 42.7%, mientras que los contratos de dos años presentan un riesgo residual de apenas el 2.8%.

Impacto de los Servicios: Los usuarios de Fibra Óptica presentan una tasa de abandono alarmante del 41.9%, a pesar de ser un producto de alto valor. Por el contrario, los servicios adicionales como soporte técnico actúan como "anclas de retención"; los clientes sin este servicio desertan 2.7 veces más.

Perfil Financiero: El Churn afecta desproporcionadamente a los clientes con cargos mensuales altos (mediana de 80.0 en desertores vs. 63.0 en retenidos), lo que indica que la empresa está perdiendo a sus clientes más rentables.

**Conclusiones e Insights**

Tras el análisis exhaustivo de los documentos, se extraen las siguientes conclusiones clave:

La Crisis de los 6 Meses: Más del 53% del abandono ocurre en el primer semestre de vida del cliente. Esto sugiere que el problema principal no es solo el precio, sino la experiencia temprana (instalación, soporte inicial y cumplimiento de expectativas).

El Punto de Inflexión de los 29 Meses: Este es un límite crítico donde el ciclo de vida se vuelve vulnerable, coincidiendo con el vencimiento de contratos estándar y la necesidad de renovación.

Estabilidad Familiar: El 48.4% de los clientes tiene pareja y el 30% tiene dependientes. Se observó que los clientes con vínculos familiares son significativamente más estables (19.7% de churn) que los clientes solteros (33% de churn).

Fricción Operativa en Pagos: El uso de cheque electrónico está ligado a una tasa de churn del 45.3%, lo que sugiere problemas en la pasarela de pagos o una falta de compromiso del cliente con el método manual.

**Recomendaciones**

Para mitigar la evasión, se proponen las siguientes líneas estratégicas:

Programa de Onboarding Intensivo: Diseñar un seguimiento proactivo durante los primeros 90 días, con especial enfoque en los usuarios de Fibra Óptica para asegurar una transición técnica exitosa.

Incentivos de Migración Contractual: Implementar campañas para convertir a los clientes "mes a mes" a contratos anuales mediante descuentos estratégicos, atacando directamente el foco más alto de deserción.

Promoción de Planes Familiares y Convergentes: Aprovechar la estabilidad del segmento familiar creando paquetes multicompartidos que agrupen líneas y servicios, elevando el costo de migración para el usuario.

Refuerzo de Servicios de Valor Agregado: Integrar servicios de "seguridad online" y "soporte técnico" en los planes básicos. Al ser servicios que generan dependencia positiva, reducen drásticamente la probabilidad de cancelación.

Alertas de Retención Preventiva: Utilizar los datos de antigüedad para lanzar ofertas de renovación antes de que el cliente alcance los 29 meses o el vencimiento de su contrato actual.
